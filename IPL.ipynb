Here, we are performing an analysis of Top 3 teams per season that have the overall high win ratio. Creating a custom function for the same

```python
mi_df = df[(df['Team1'] == 'Mumbai Indians') | (df['Team2'] == 'Mumbai Indians')]
```

```python
mi_df
```

```python
mi_df.groupby('Season')['ID'].count()
```

```python
dict_ = {
    'Season': [],
    'Winning Ratio': []
}
```

```python
mi_s = mi_df.groupby('Season')
```

```python
for group,data in mi_s:
  dict_['Season'].append(group)
  dict_['Winning Ratio'].append(round(data[data['WinningTeam'] == 'Mumbai Indians'].shape[0]/data.shape[0],2))
```

```python
pd.DataFrame(dict_)
```

```python
#lets make a general function
def findWinRatio(name):

  dict_ = {
    'Season': [],
    'Winning Ratio': []
  }

  temp_groups = df[(df['Team1'] == name) | (df['Team2'] == name)].groupby('Season')

  for group,data in temp_groups:
    dict_['Season'].append(group)
    dict_['Winning Ratio'].append(round(data[data['WinningTeam'] == name].shape[0]/data.shape[0],2))

  temp_df = pd.DataFrame(dict_)
  plt.title(f"Winning ratio of {name}")
  sns.barplot(temp_df,x='Winning Ratio',y='Season',orient='h',hue='Season')
  plt.xlabel('Winning %')
```

```python
findWinRatio('Mumbai Indians')
```

### Q) Calculate the total runs made by the Teams during the powerplay overs in IPL 2017

```python
# We will calculate average runs made by each team during their powerplay overs in IPL 2017
tf_d3 = matches.merge(delivery,left_on='id',right_on='match_id')
```

```python
temp_res = tf_d3[(tf_d3['season'] == 2017) & (tf_d3['over'] <= 6)].groupby('batting_team')['total_runs'].sum()
```

```python
temp_res
```

```python
#Total runs made by teams during the powerplay overs in IPL 2017
res6 = temp_res.reset_index().sort_values(by=['total_runs'],ascending=False)
```

```python
res6
```

```python
# plt.figure(figsize=(8,9))
plt.title("Total PowerPlay Runs")
a = sns.barplot(res6,x='total_runs',y='batting_team')
for i in a.containers:
  a.bar_label(i)
plt.xlabel('Total Runs')
plt.ylabel('Team')
```

### Analysis: KKR have made the highest runs during the powerplay overs followed by MI while RCB recorded the lowest score of all teams

```python
inter = matches[matches['season'] == 2017]
```

```python
total_matches_played = inter.groupby('team1')['team1'].count() + inter.groupby('team2')['team2'].count()
```

```python
total_matches_played
```

```python
res7 = (round(temp_res / total_matches_played,2)).reset_index().rename(columns={0:'Avg Runs'})
```

```python
#Average runs made in the powerplay
res7
```

```python
# plt.figure(figsize=(8,9))
plt.title("Avg Powerplay Runs in IPL 2017")
a = sns.barplot(res7,x='Avg Runs',y='batting_team')
for i in a.containers:
  a.bar_label(i)
plt.xlabel('Avg Powerplay Runs')
plt.ylabel('Team')
```

### Analysis: KKR have scored 56 runs during the powerplay overs on an average followed by Gujarat Lions with 54 runs and RCB with 41 runs

```python
#we can define a custom function for the above analysis of powerplay runs
def getPowerPlayRuns(season):
  int_1 = tf_d3[(tf_d3['season'] == season) & (tf_d3['over'] <= 6)].groupby('batting_team')['total_runs'].sum()
  inter = matches[matches['season'] == season]
  total_matches_played = inter.groupby('team1')['team1'].count() + inter.groupby('team2')['team2'].count()
  return (round(temp_res / total_matches_played,2)).reset_index().rename(columns={0:'Avg Runs'})
```

```python
getPowerPlayRuns(2017)
```
